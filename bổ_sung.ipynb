```python
# 3 Kiểm tra số chẵn hay lẻ
a = int(input("Nhập một số nguyên: "))
if a % 2 ==0:
  print(a, "là số chẵn")
else:
  print(a, "là số lẻ")
```

```python
# 4 tìm số lớn nhất trong 3 số
a = int(input("Nhập số thứ nhất: "))
b = int(input("Nhập số thứ hai: "))
c = int(input("Nhập số thứ ba: 3"))
so_max = max(a, b, c)
print("số lớn nhất là:", so_max)
```

```python
# 5 Kiểm tra năm nhuận
năm = int(input("Nhập năm: "))
if (năm % 4 == 0 and năm % 100 !=0 or năm % 400 == 0):
  print(năm, "là năm nhuận")
else:
  print(năm, "là năm không nhuận")
```

```python
# 6 Tính tổng các số từ 1 đến n
n = int(input("Nhập số nguyên dương n:  "))
tổng = 0
for i in range(1, n+1):
    tổng += i
print("tổng từ 1 đến n ", n, "là : ", tổng)
```

```python
# 7  Đếm số nguyên âm, dương, bằng 0
ds = []
print("Nhập 10 số nguyên : ")
for i in range(10):
  số = int(input("số thứ {i+1}: "))
  ds.append(số)
âm = 0
dương = 0
bang0 = 0
for x in ds:
    if x < 0:
        âm += 1
    elif x > 0:
        dương += 1
    else:
      bang0 += 1
print("số âm :" , âm)
print("số dương::" , dương)
print("số bằng 0 :" , bang0)
```

```python
# 8 Bảng cửu chương
n = int(input(" Nhập 1 số từ 1 tới 9 : "))
print("Bảng cửu chương của {n}:")
for i in range(1, 11):
  kết_quả = n * i
  print({n}, "x" ,{i}, "=", {kết_quả} )
```

```python
# 9 Kiểm tra số nguyên tố
n = int(input("Nhập số nguyên dương n: "))
if n < 2:
    print(n, "không phải số nguyên tố")
else:
  ngto = True
  for i in range(2, int(n**0.5) + 1):
      if n % i == 0 :
         ngto = False
         break
  if ngto:
    print(n, "là số nguyên tố ")
  else:
    print(n,"không phải là số nguyên tố")
```

```python
# 10 Đảo ngược chuỗi
s = input("Nhập một chuỗi: ")
dao_nguoc = ""
for i in range(len(s)-1, -1, -1):
    dao_nguoc += s[i]
print("Chuỗi đảo ngược là:", dao_nguoc)
```
